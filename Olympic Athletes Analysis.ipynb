# Data Cleaning

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['nationality', 'country', 'weight', 'nationality_code', 'hero', 'occupation', 'reason', 'ritual', 'hobbies', 'residence_place'], inplace = True )
df.drop(columns = ['sporting_relatives', 'height', 'name_short', 'name_tv'], inplace =True)
```

```python
df.drop(columns = ['philosophy', 'influence'], inplace =True)
```

```python
df['disciplines']= df['disciplines'].str.strip("['']123._/")
```

```python
df['events']= df['events'].str.strip("['']123._/")
```

```python
df['events']= df['events'].str.strip('""')
```

```python
df['name'] = df['name'].str.strip()
```

```python
df['name']= df['name'].apply(lambda x: x.title())
```

```python
df = df.fillna(' ')
```

```python
df['birth_place']= df['birth_place'].apply(lambda x: x.title())
```

```python
df['sport'] = df['disciplines'] 
```

```python
df.info()
```

```python
df['birth_date'] = pd.to_datetime(df['birth_date'])
```

```python
df.info()
```

```python
df
```

```python
df['coach'] = df['coach'].str.replace('<br>', '')
```

```python
df[['sports', 'sport1']]=df['disciplines'].str.split(',', n=1, expand=True)
df[['sport2', 'sport3']]=df['sports'].str.split("'", n=1, expand=True)
```

```python
df['sport1']=df['sport1'].fillna('')
df['sport2']=df['sport2'].fillna('')
df['sport3']=df['sport3'].fillna('')
```

```python
a= df['sport2'].value_counts()
b = df['sport1'].value_counts()
print(f'{a}-> {b}')
```

```python
df.drop(columns =['sports', 'sport1', 'sport', 'sport3'], inplace = True)
```
